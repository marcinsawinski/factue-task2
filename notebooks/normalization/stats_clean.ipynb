# cleanup


```python

root = Path("data/llm_output/normalization/cleanup")
# Example path template
files = root.rglob("**/*.parquet")

# Read all files and add file path as a column
df_list = []
for f in files:
    df_part = pd.read_parquet(f)
    df_part['source_file'] = f  # add the file path
    df_list.append(df_part)

# Combine into one DataFrame
df = pd.concat(df_list, ignore_index=True)
df['split'] = df.source_file.astype(str).str.split('/',expand=True)[7]
df['init_part_no'] = df.source_file.astype(str).str.split('/',expand=True)[9].str.split('_',expand=True)[1].str.split('.',expand=True)[0].astype(int)
df['lang'] = df.source_file.astype(str).str.split('/',expand=True)[8]
df['split'].value_counts()
```

```python
clean_cols = ['text', 'reference', 'split', 'lang','init_part_no']

cleaned = df[(df.check_result==0)&(df.text_lang==df.reference_lang)].sort_values('error')
cleaned = cleaned[clean_cols].sample(frac=1, random_state=42).sort_values(['split','lang'])
cleaned
```

```python
batch_size = 10

for (split, lang), group in cleaned.groupby(['split', 'lang']):
    dir_path = Path(f"data/cleaned/normalization/{split}/{lang}")
    dir_path.mkdir(parents=True, exist_ok=True)
    # Split indices into batches, then select rows by index
    indices_batches = np.array_split(group.index, max(1, len(group) // batch_size + (len(group) % batch_size > 0)))
    for i, idx in enumerate(indices_batches):
        batch = group.loc[idx]
        file_path = dir_path / f"{lang}_part_{i:04d}.parquet"
        batch.to_parquet(file_path, index=False)
        print(f"saved: {file_path}")
```
