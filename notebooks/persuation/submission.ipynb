# llm outputs

```python
def normalize_binary(x):
    return 1 if str(x).strip().lower() in {'1', '1.0','true'} else 0
def load_outputs(model_name, prompt_version):
    root = Path("data/llm_output/persuasion/detect")
    # Example path template
    files = root.rglob(f"{model_name}/{prompt_version}/*/test/*/*.parquet")

    # Read all files and add file path as a column
    df_list = []
    for f in files:
        df_part = pd.read_parquet(f)
        df_part['source_file'] = f  # add the file path
        df_list.append(df_part)

    # Combine into one DataFrame
    df = pd.concat(df_list, ignore_index=True)
    df['pred'] = df['pred'].apply(normalize_binary)
    return df
```

# output struct

```python
from pathlib import Path
import pandas as pd
from factue.utils.vars import PROJECT_ROOT
import numpy as np

base_dir = PROJECT_ROOT / "data/raw/persuasion/test_data"
langs = ["PL", "RU", "BG", "SI", 'HR']

records = []

for lang in langs:
    lang_path = base_dir / lang
    raw_docs_path = lang_path / "raw-documents"
    input_path = lang_path / "input-file.txt"

    # Parse subtask-1 annotations and extract spans
    with input_path.open(encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split("\t")
            if len(parts) != 3:
                continue
            fname, start, end = parts
            start, end = int(start), int(end)


            records.append(
                {
                    "filename": fname,
                    "start": start,
                    "end": end,
                    "text_lang": lang,
                }
            )

# Create a DataFrame
output_keys = pd.DataFrame(records)

output_keys
```

# Convert

```python
df = df.assign(detected=np.where(df['pred'] == 1, df['prompt_name'], None))
df
```

```python
detected = df[df['detected'].notna()].groupby(['filename', 'start', 'end', 'text_lang'])['detected'].agg(list).reset_index()
detected
```

```python
output_data = output_keys.merge(detected, on=['filename','start','end','text_lang'], how='left')
output_data
```

```python
for lang, group in output_data.groupby('text_lang'):
    with open(f"data/submit/subtask2_{lang.upper()}_{str(submission_version)}.csv", "w", encoding="utf-8") as f:
        for _, row in group.iterrows():
            base_info = f"{row['filename']}\t{row['start']}\t{row['end']}"
            if isinstance(row['detected'], list):
                detected_items = "\t".join(map(str, row['detected']))
                base_info = f"{base_info}\t{detected_items}"
            f.write(f"{base_info}\n")
```
