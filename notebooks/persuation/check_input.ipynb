```python
import os

def get_all_files(base_dir):
    file_set = set()
    for root, _, files in os.walk(base_dir):
        for file in files:
            # Get path relative to base_dir
            rel_path = os.path.relpath(os.path.join(root, file), base_dir)
            file_set.add(rel_path)
    return file_set

# Replace these with your folder paths
folder1 = 'data/raw/persuasion/trial_data/TRIAL-ANNOTATED'
folder2 = 'data/raw/persuasion/train_data'

files1 = get_all_files(folder1)
files2 = get_all_files(folder2)

# Find files in folder1 but not in folder2
missing_in_folder2 = files1 - files2

# Print or save the result
print("Files present in folder1 but missing in folder2:")
for file in sorted(missing_in_folder2):
    print(file)

for file in sorted(files1):
    print(file, file in files2)
```

```python
import pandas as pd
from pathlib import Path
import os
import numpy as np
root = Path("data/preprocessed/persuasion/")
# Example path template
files = root.rglob("**/*.parquet")

# Read all files and add file path as a column
df_list = []
for f in files:
    df_part = pd.read_parquet(f)
    df_part['source_file'] = f  # add the file path
    df_list.append(df_part)

# Combine into one DataFrame
df = pd.concat(df_list, ignore_index=True)
df.sort_values(['filename','start'])
```
