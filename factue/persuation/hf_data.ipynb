```python
from datasets import load_from_disk
from pathlib import Path
from factue.utils.vars import project_root
import pandas as pd
# Path to your saved .hf dataset folder
dataset_path = project_root / "data/raw/persuasion/propaganda.hf"

# Load the dataset
dataset = load_from_disk(dataset_path)

# Explore the dataset
print(dataset)
print(dataset['tw_pl_de'])  # or dataset['train'][0] if it's a DatasetDict
```

```python
dfs = []
for split_name, split_data in dataset.items():
    df = split_data.to_pandas()
    df["split"] = split_name
    dfs.append(df)
# Concatenate all splits into one DataFrame
full_df = pd.concat(dfs, ignore_index=True)

# Show the combined DataFrame
full_df.sort_values(by=["filename", 'start'])
```
